### 3-1. 統計値を Simulation してみよう

#### ２年分の Data から１ヶ月分の Data を抽出する Simulation

```python
# 正規分布を生成する
import numpy as np
import matplotlib.pyplot as plt

# 母集合の大きさを設定
num = 365 * 2  # ２年間に相当する(365日 × 2)

# 乱数の平均・標準偏差を設定
ave = 0.0
std = 1.0

# 乱数を発生（Seed を固定）
np.random.seed(seed=0)
x = np.random.normal(ave, std, num)  # 正規乱数を発生

# 平均・標準偏差を計算
x_ave = np.average(x)
x_std = np.std(x)
print('平均: ', x_ave)
print('標準偏差: ', x_std)

# 描画
num_bin = 21
plt.hist(x, num_bin, color='k')
plt.xlim([-5, 5])
plt.show()
```

```python
# Random-sampling した標本の平均の分布を描画する
import numpy as np

# Sample数（Sample集団の大きさ）を設定
num_sample = 30  # １ヶ月間に相当する Sample数を設定

# 試行回数を設定
num_trial = 10_000
x_trial = np.zeros(num_trial)

# Sample平均の算出を試行
for i in range(num_trial):
    x_sample = np.random.choice(x, num_sample)  # Sample を抽出
    x_ave = np.average(x_sample)  # 平均を計算
    x_trial[i] = x_ave  # 値を格納

# Sample平均の平均・標準偏差を計算
x_trial_ave = np.average(x_trial)
x_trial_std = np.std(x_trial)
print('平均: ', x_trial_ave)
print('標準偏差:', x_trial_std)

# 描画
num_bin = 21
plt.hist(x_trial, num_bin, color='k')
plt.xlim([-5, 5])
plt.show()
```

ほぼ、母集団の平均値と同じ。標準偏差は母集団の標準偏差に比べて小さい。

### 3-2.中心極限定理について知ろう
**中心極限定理** が正しいか Source-code を実行して確かめてみる。
> #### 中心極限定理
> 母集団が
>    - 平均 µ
>    - 標準偏差 σ
>
> に従う場合、標本平均は µ。
>
> 標準偏差 σ / √n の正規分布に従う（n は標本集合の Size.）

```python
# 中心極限定理を用いて標本集合の標準偏差を求める
import numpy as np

org_std = 1.0  # 母集団の分散を設定
num_sample = 30  # Sample集合の大きさを設定
sample_std = org_std / np.sqrt(num_sample)  # Sample集合の平均の標準偏差を計算
print(f"Sample集合の平均の標準偏差: {sample_std}")
```

前節で実行した標本集合の標準偏差である0.18 とほぼ同じ。中心極限定理によって標本平均のばらつきをうまく計算できることがわかる。

### 3-3. １ケ月の Data を正確に取ってみよう
Hotel の Data を用いながら、盗難状況について分析していく。

```python
# 1ヶ月分の被害List を読み込む
import pandas as pd

df_theft_201811 = pd.read_csv('theft_list_201811.csv', index_col=0, parse_dates=[0])
df_theft_201811
```

```python
# Amenity の金額List を読み込む

df_amenity_price = pd.read_csv('amenity_price.csv', index_col=0)
df_amenity_price
```

```python
# １ヶ月分の被害List から被害総額を計算する
total_amount = 0
total_theft = 0
for i_index in range(len(df_theft_201811.index)):
    for i_column in range(len(df_theft_201811.columns)):
        total_amount += df_theft_201811.iloc[i_index, i_column] * df_amenity_price['金額'].iloc[i_column]
        total_theft += df_theft_201811.iloc[i_index, i_column]
        if df_theft_201811.iloc[i_index, i_column] >= 0:
            print(df_theft_201811.index[i_index], df_theft_201811.columns[i_column],
                  df_theft_201811.iloc[i_index, i_column], "点")

print(f"被害総額: {total_amount} 円")
print(f"被害件数: {total_theft} 件")
```

2018年11月の被害状況
- １日２件以上の被害が発生している。
- 総額が決して少なくない。

この数字にどの程度のばらつきがあるのかを事前に確認し、１～２年間の被害総額を概算として見積、対策につなげていく。
