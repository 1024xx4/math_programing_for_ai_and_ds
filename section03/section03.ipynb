### 3-2.中心極限定理について知ろう
**中心極限定理** が正しいか Source-code を実行して確かめてみる。
> #### 中心極限定理
> 母集団が
>    - 平均 µ
>    - 標準偏差 σ
>
> に従う場合、標本平均は µ。
>
> 標準偏差 σ / √n の正規分布に従う（n は標本集合の Size.）

```python
# 中心極限定理を用いて標本集合の標準偏差を求める
import numpy as np

org_std = 1.0  # 母集団の分散を設定
num_sample = 30  # Sample集合の大きさを設定
sample_std = org_std / np.sqrt(num_sample)  # Sample集合の平均の標準偏差を計算
print(f"Sample集合の平均の標準偏差: {sample_std}")
```

前節で実行した標本集合の標準偏差である0.18 とほぼ同じ。中心極限定理によって標本平均のばらつきをうまく計算できることがわかる。

### 3-4. １ヶ月分の Data から２年分の Data の平均・標準偏差を推定しよう
前節の１ヶ月分の Data の平均と標準偏差を求めることで２年間の Data の平均と標準偏差を、中心極限定理が成り立つことを前提に計算してみる。

```python
# １日ごとの被害額を List化する
import numpy as np
import matplotlib.pyplot as plt

list_amount = np.zeros(len(df_theft_201811.index))
for i_index in range(len(df_theft_201811.index)):
    for i_column in range(len(df_theft_201811.columns)):
        list_amount[i_index] += df_theft_201811.iloc[i_index, i_column] * df_amenity_price['金額'].iloc[i_column]
plt.plot(list_amount, color='k')
plt.show()
```

１日ごとの被害額を List化した結果を Graph化。
第４日目の被害額が100,000円と突出している。原因は、TV の盗難事故が発生していることによる結果。
一旦、例外とせずにこのまま進める。

```python
# 10日間を Random-sampling して平均値の分布を算出する
import numpy as np

num_sample = 10  # Sample数（Sample集団の大きさ）を設定

# 試行回数を設定
num_trial = 10_000
x_trial = np.zeros(num_trial)

# Sample平均の算出を試行
for i in range(num_trial):
    # Sample抽出
    x = list_amount
    x_sample = np.random.choice(x, num_sample)
    x_ave = np.average(x_sample)  # 平均を計算
    x_trial[i] = x_ave  # 値を格納

# Sample平均の平均・標準偏差を計算
x_trial_ave = np.average(x_trial)
x_trial_std = np.std(x_trial)
print(f"平均: {x_trial_ave}")
print(f"標準偏差: {x_trial_std}")

# 描画
num_bin = 21
plt.hist(x_trial, num_bin, color='k')
plt.xlim([-50_000, 50_000])
plt.show()
```

１ヶ月分の Data から１日ごとの被害額を List化し、そのうち10日間を Random-sampling。その平均値の分布を求めた。
第４日目の被害額が突出していることで中心極限定理が教える正規分布とはほど遠い形状になっている。
しかし、それでも求めた平均値と標準偏差の値を利用して、母集団である２年間の Data の標準偏差を逆推定してみる。

```python
# １ヶ月分の Data から２年分の Data の平均・標準偏差を推定する
import numpy as np

sample_std = x_trial_std.round()  # Sample集合の平均の標準偏差を推定
org_std = np.sqrt(num_sample) * sample_std  # 母集団の分散を計算
print(f"母集団の標準偏差: {org_std}")
```

Random-sampling から算出した標準偏差から、母集団である２年間の Data の標準偏差を逆推定（Sample数の平方根に標準偏差を乗算）。
結果として、平均値に対して、標準偏差がかなり大きな値となった。
それだけ、標準集合のばらつきが大きいということになる。

### 3-5. 標準偏差と信頼度の関係を知ろう
前節で算出した平均値と標準偏差値の持つ意味を**信頼度**から考えていく。

```python
# 標準偏差から信頼度を計算する
ratio = 1.0  # 標準偏差の倍率の設定
x_trial_out1 = x_trial[x_trial > x_trial_ave + ratio * x_trial_std]  # 左側の領域外の計算
x_trial_out2 = x_trial[x_trial < x_trial_ave - ratio * x_trial_std]  # 右側の領域外の計算
reliability = 1 - (len(x_trial_out1) / len(x_trial) + len(x_trial_out2) / len(x_trial))  # 信頼度の計算
print(f"信頼度: {reliability}")
```

### 3-6. 宿泊者数との相関関係を仮定して被害額の推移を推測しよう
宿泊者１人あたりの平均被害額を算出し、２年間の１日ごとの顧客者数から２年間の推定被害額を計算してみる。
##### 理由
- １日あたりの宿泊者数と被害額が比例すると仮定。
- １日あたりの被害額を推定するうえで十分に信頼できる区間を考えると、かなり大きな幅が必要となりそうだが、２年間の合計として考えると被害総額の推定自体は、それなりに正確に行なえるかも知れない。その推定をどの程度正確に行うことができるのかを考える土台にしたい。

```python
# 宿泊者１人あたりの平均被害額を算出する
import pandas as pd
import datetime as dt

theft_per_day = x_trial_ave.round()  # １日あたりの被害額の平均値を設定
df_into = pd.read_csv('accomodation_info.csv', index_col=0, parse_dates=[0])  # 宿泊Data 読み込み

# １日あたりの宿泊者数を抽出
x = df_into.resample('D').count()
df_num = x.iloc[:, 0]

# １ヶ月分の宿泊者数を抽出
target_date = dt.datetime(2018, 11, 30)
df_num_201811 = df_num[df_num.index <= target_date]
print(f"１ケ月の宿泊者数: {sum(df_num_201811)}")

# 1ヶ月の宿泊者数から１日あたりの平均宿泊者数を計算
num_per_day = sum(df_num_201811) / len(df_num_201811)
print(f"１日あたりの平均宿泊者数: {num_per_day}")

# 宿泊者１人あたりの平均被害額
theft_per_person = theft_per_day / num_per_day
print(f"宿泊者数１人あたりの平均被害額: {theft_per_person}")
```

```python
# ２年間の被害額の推移を推定する
import numpy as np

estimated_theft = np.zeros(len(df_num))
for i in range(len(df_num)):
    estimated_theft[i] = df_num.iloc[i] * theft_per_person
df_estimated_theft = pd.DataFrame(estimated_theft, index=df_num.index, columns=['推定被害額'])
print(f"２年間の推定被害総額: {sum(df_estimated_theft['推定被害額'])}")
plt.plot(df_estimated_theft, color='k')
plt.xticks(rotation=60)
plt.show()
```

### 3-7. 年間の被害総額とその信頼区間を推定しよう
推定した２年間の被害額の推移に対して、母集団の標準偏差から宿泊者１人あたりの標準偏差を割り出し、その信頼区間を描画してみる。

```python
# ２年間の被害額の信頼区間を推定する
import matplotlib.pyplot as plt

# 標準偏差の設定
theft_std_per_day = org_std.round()
theft_std_per_person = theft_std_per_day / num_per_day
print(f"宿泊者１人あたりの平均被害総額の標準偏差: {theft_std_per_person}")

# 信頼区間の設定
list_estimated_theft = []
for i in range(len(df_num)):
    temp_ave = df_num.iloc[i] * theft_per_person
    temp_std = df_num.iloc[i] * theft_std_per_person
    temp = [temp_ave - temp_std, temp_ave, temp_ave + temp_std]
    list_estimated_theft.append(temp)

# 描画
plt.boxplot(list_estimated_theft)
plt.xticks(color='None')
plt.show()
```
