## 7-7. 人間関係の Network を可視化してみよう

```python
# Link-data を読み込む#
import pandas as pd

df_links = pd.read_csv("links.csv", index_col=0)
df_links
```

```python
# Graph を描画する
import networkx as nx
import matplotlib.pyplot as plt

# Graph-object の作成
G = nx.Graph()  # Network を描画することを宣言

# 頂点の設定
# Link-data として読み込んだ Node を生成
NUM = len(df_links.index)
for i in range(0, NUM):
    node_no = df_links.columns[i].strip("Node")
    G.add_node(str(node_no))

# 辺の設定
# Link-data において「1」と記載されている Node間をつなぐ Link を生成。
for i in range(NUM):
    for j in range(NUM):
        if df_links.iloc[i][j] == 1:
            G.add_edge(str(i), str(j))

# 描画
plt.figure(figsize=(12, 8))
nx.draw_networkx(G, node_color="k", edge_color="k", font_color="w")
plt.show()
```

## 7-8. 人間関係の Network の成長の様子を可視化してみよう
初期User の中から **「Hub」** と呼ばれる中心的な役割を担う存在が現われて、SNS などの Network は成長する。

```python
# Link-data を読み込む
import pandas as pd

df_links = pd.read_csv("links.csv", index_col=0)
```

```python
# Node を追加する
import numpy as np

N_plus = 100  # 追加する Node数
N = len(df_links.index)
for i in range(N, N + N_plus):  # 追加する Node数分だけ Node追加の処理を行なう。
    # 接続する Node を決定
    j = int(np.random.rand() * (i - 1))  # すでに存在する Node の中から Random に１つを選択
    node_name_i = "Node" + str(i)
    node_name_j = "Node" + str(j)
    # 列を追加
    df_links[node_name_i] = 0
    # 行を追加
    list_zero = [[0] * (len(df_links.index) + 1)]
    s = pd.DataFrame(list_zero, columns=df_links.columns.values.tolist(), index=[node_name_i])
    df_links = pd.concat([df_links, s])
    # Link を追加
    df_links.loc[node_name_i, node_name_j] = 1  # 選択した接続先である Node に対応するところに Link が接続されている意味の「1」を書き込む
    df_links.loc[node_name_j, node_name_i] = 1  # 選択した接続先である Node に対応するところに Link が接続されている意味の「1」を書き込む
df_links
```

```python
# Graph を描画する
import networkx as nx
import matplotlib.pyplot as plt

# Graph-object の作成
G = nx.Graph()

# 頂点の設定
NUM = len(df_links.index)
for i in range(0, NUM):
    node_no = df_links.columns[i].strip("Node")
    G.add_node(str(node_no))

# 辺の設定
for i in range(NUM):
    for j in range(NUM):
        if df_links.iloc[i][j] == 1:
            G.add_edge(str(i), str(j))

# 描画
plt.figure(figsize=(12, 8))
nx.draw_networkx(G, node_color="k", edge_color="k", font_color="w")
plt.show()
```

中心に近い位置に配置されている Node に多くの Node からの Link が集中している。これが、**Hub** と呼ばれる中心的な存在を表す。
初期状態にこれらの村税が多くの Node に対して働きかを行ない、Link を増やしていき、Network全体が成長しく。

### 7-9. Network を分析してみよう
生成された Network がどのような統計的性質を持つのか明らかにするために、Network の各Node が持つ Link数の分布を確認する。

```python
# Node ごとの Link数を描画する
import numpy as np
import matplotlib.pyplot as plt

# Link数を Count
list_nodenum = np.zeros(len(df_links.index))
for i in range(len(df_links.index)):
    node_name_i = "Node" + str(i)
    list_nodenum[i] = sum(df_links[node_name_i].values)
plt.bar(range(len(df_links.index)), list_nodenum)
plt.show()
```

初期の Node に Link数が集中しているものの、必ずしもすべての Node が平等に Link を獲得しているわけではなく、**「偏り」** が見える。

```python
# Histogram を描画する
plt.hist(list_nodenum)
plt.show()
```

べき分布に近いものであることがわかり、少ない **「Hub」** となる人が多くの Link を独占する状況が確認とれる。

## 7-10. 微分方程式を差分化する際の誤差とその対策について知っておこう

```python
# ねずみ算の Parameter: dt と num の関係
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter設定
dt = 1.0  # ねずみが生まれ、次に子孫を残す(delta の演算を行なう)際の時間間隔
a = 1.0
num = 10
```

```python
# 初期化（初期値設定）
n = np.zeros(num)
t = np.zeros(num)
n[0] = 2.0
t[0] = 0.0
```

```python
# 時間発展方程式
for i in range(1, num):
    t[i] = t[i - 1] + dt
    delta = a * n[i - 1]
    n[i] = delta * dt + n[i - 1]

# Graph描画
plt.plot(t, n, color="blue")
plt.show()
```

```python
# Parameter設定
dt = 0.1  # ねずみが生まれ、次に子孫を残す(delta の演算を行なう)際の時間間隔
a = 1.0
num = 100  # 総時間
```

```python
# 初期化（初期値設定）
n = np.zeros(num)
n[0] = 2.0
t = np.zeros(num)
t[0] = 0.0
```

```python
# 時間発展方程式
for i in range(1, num):
    t[i] = t[i - 1] + dt
    delta = a * n[i - 1]
    n[i] = delta * dt + n[i - 1]

# Graph描画
plt.plot(t, n)
plt.show()
```

微分方程式を数式によって解くことで、本来解きたかった方程式を導くことを **「微分方程式を解析的に解く」** と言う。
実際には、複雑な微分方程式を解析的に解くことは難しいため、数値Simulation によって **「近似解」** を導くことになる。

#### RungeーKutta法
数値Simulation による微分方程式の解の誤差を小さくする方法。
Euler法は、微分方程式を手軽に実装するには適しているが、解析解との誤差が大きくなるという問題点がある為、なるべく厳密に解を求めたい場合には **「Runge-Kutta法」** を用いる。

```python
# ねずみ算を Euler法によって解く
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter設定
dt = 0.1
a = 1.0
T = 10
num = int(T / dt)
```

```python
# 初期化（初期値設定）
n = np.zeros(num)
t = np.zeros(num)
n[0] = 2.0
t[0] = 0.9
```

```python
# 時間発展方程式
for i in range(1, num):
    t[i] = t[i - 1] + dt
    delta = a * n[i - 1]
    n[i] = delta * dt + n[i - 1]

# Graph描画
plt.plot(t, n, color="blue")
plt.show()
```

```python
# ねずみ算術の Euler法による解と解析解との比較を行なう
t = np.arange(0, T, dt)
n_cont = n[0] * np.exp(a * t)
print(len(n_cont), len(n))
plt.plot(t, n)
plt.plot(t, n_cont, color="red")
plt.show()
```

```python
# ねずみ算を Runge-Kutta法によって解く
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter設定
dt = 0.1
a = 1.0
T = 10
num = int(T / dt)
```

```python
# 初期化（初期値設定）
n_runge_kutta = np.zeros(num)
t = np.zeros(num)
n_runge_kutta[0] = 2.0
t[0] = 0.0
```

```python
# 時間発展方程式を定める関数
def f(n, t):
    return n


# 時間発展方程式
for i in range(1, num):
    t[i] = t[i - 1] + dt
    k1 = dt * f(n_runge_kutta[i - 1], t[i - 1])
    k2 = dt * f(n_runge_kutta[i - 1] + k1 / 2, t[i - 1] + dt / 2)
    k3 = dt * f(n_runge_kutta[i - 1] + k2 / 2, t[i - 1] + dt / 2)
    k4 = dt * f(n_runge_kutta[i - 1] + k3, t[i - 1] + dt)
    n_runge_kutta[i] = n_runge_kutta[i - 1] + 1 / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
```

```python
# Graph描画
plt.plot(t, n_runge_kutta, color="green")
plt.show()
```

```python
# ねずみ算の Euler法・Runge-Kutta法による解と解析解を比較する
t = np.arange(0, T, dt)
n_cont = n[0] * np.exp(a * t)
print(len(n_cont), len(n))
plt.plot(t, n, linewidth=4, color="blue")
plt.plot(t, n_runge_kutta, linewidth=4, linestyle="dashed", color="green")
plt.show()
```

Runge-Kutta法によって求めた解が、解析解に比べてほとんど誤差がない、ということがわかる。
